```python
import os
import random
import shutil
```

```python

splitsize = .85
categories = []
```

```python

source_folder= 'C:/Users/carlo/OneDrive/Documentos/Python/PythonJupyterPy310/MachineLeaning/Atividade/Vehicles'
folders = os.listdir(source_folder)
print(folders)
```

```python
#Aqui é onde ele vai pega 2 classes (Carro e Moto)
for subfolder in folders:
    if os.path.isdir(source_folder + '/' + subfolder):
        categories.append(subfolder)

categories.sort()
print(categories)
```

```python
#criação da pasta onde vai ser armazenado o dataset de Treino e Validação

target_folder = 'C:/Users/carlo/OneDrive/Documentos/Python/PythonJupyterPy310/MachineLeaning/Atividade'
existDataSetPath = os.path.exists(target_folder) #Uma variavel que recebe o caminho do diretorio

if existDataSetPath == False:  
    os.mkdir(target_folder)

#Print que mostra o diretorio do arquivo
print (target_folder) 
```

```python
def split_data(SOURCE,TRAINING,VALIDATION,SPLIT_SIZE):
    files=[]

    for filename in os.listdir(SOURCE):
        file = SOURCE + filename
        print (file)
        if os.path.getsize(file) > 0:
            files.append(filename)
        else:
            print (filename + ' tamanho é zero, ignore isso...')
    print (len(files))
    trainingLength = int(len(files) * SPLIT_SIZE)
    shuffleSet = random.sample(files,len(files))
    trainingSet = shuffleSet[0:trainingLength]
    validSet = shuffleSet[trainingLength:]

    #Copia os treinos das imagens
    for filename in trainingSet:
        thisFile = SOURCE + filename
        destination = TRAINING + filename
        shutil.copyfile(thisFile, destination)

    #Copia as validações das imagens
    for filename in validSet:
        thisFile = SOURCE + filename
        destination = VALIDATION + filename
        shutil.copyfile(thisFile, destination)
```

```python
trainPath = target_folder + '/train'
validatePath = target_folder + '/validate'

#criação de pasta de target

existDataSetPth = os.path.exists(trainPath)

if existDataSetPth == False:
    os.mkdir(trainPath)

existDataSetPth = os.path.exists(validatePath)

if existDataSetPth == False:
    os.mkdir(validatePath)
```

```python
#para rodar as funções dentro dos arquivos selecionados

for category in categories:
    trainDestPth = trainPath + '/' + category
    validateDestPth = validatePath + '/' + category

    if os.path.exists(trainDestPth)== False:
        os.mkdir(trainDestPth)
    if os.path.exists(validateDestPth)==False:
        os.mkdir(validateDestPth)

    sourePath = source_folder + '/' + category + '/'
    trainDestPth = trainDestPth + '/'
    validateDestPth = validateDestPth + '/'

    print (f'Copy from : {sourePath} to : {trainDestPth} and {validateDestPth} ')

    split_data(sourePath,trainDestPth,validateDestPth,splitsize)
```
